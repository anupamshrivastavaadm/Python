##üîß Step 1: Hide the message

```python
with open('/content/sample_data/image (1).jpg', 'ab') as f:
    f.write(b'\nSecretMessage: ABCDXYZ.')
```

üß© Problem:

You used Python to embed a text message inside an

image, but assume you have a question:

"How can I tell just by looking at the image that it has

hidden text data inside?"

‚úÖ Answer:

The image will look completely normal ‚Äî you won‚Äôt see

anything with your eyes because:

- Image viewers (like Gallery, Preview, etc.) only read

  the visual part of the image.

- The data you append using open('image (1).jpg', 'ab') is

  added after the image's end marker.

- Viewers ignore anything that comes after that

  marker.

So even though the message is there, it‚Äôs invisible to the

naked eye ‚Äî but Python can still read it.

##üîç Step 2: Extract the message

```python
with open('/content/sample_data/image (1).jpg', 'rb') as f:
    data = f.read()
    if b'SecretMessage:' in data:
        print(data.split(b'SecretMessage:')[-1].decode())
```
